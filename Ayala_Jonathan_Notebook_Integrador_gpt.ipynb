* **4.2.4. Limpieza de Texto (para columnas categ√≥ricas/texto):**
    * **Columna `Entidad_Federativa`:**
        * Acciones: Eliminar espacios extra (`.str.strip()`), convertir a may√∫sculas (`.str.upper()`), estandarizar nombres de entidades si es necesario (ej. "CDMX" a "CIUDAD DE M√âXICO").
        * (Justificaci√≥n: Mejorar consistencia y permitir agrupaciones correctas.)
    * **Columna `Municipio`:**
        * Acciones: Eliminar espacios extra, convertir a may√∫sculas, corregir errores ortogr√°ficos comunes.
        * (Justificaci√≥n: Evitar duplicados por diferencias de formato y facilitar an√°lisis geogr√°fico.)
    * **Columna `Subsector`, `Categor√≠a de generador`, `Categor√≠a del residuo`, `Tipo de residuo`:**
        * Acciones: Eliminar espacios extra, convertir a may√∫sculas, estandarizar valores si es necesario.
        * (Justificaci√≥n: Mejorar calidad de los datos categ√≥ricos y facilitar an√°lisis.)

````python
# üßπ PASO 4: LIMPIEZA Y ESTANDARIZACI√ìN DE TEXTO
print("üßπ PASO 4: LIMPIEZA Y ESTANDARIZACI√ìN DE TEXTO")
print("="*50)

# Funci√≥n de estandarizaci√≥n de texto adaptada para datos delictivos
def estandarizar_texto_delictivo(texto):
    """
    Estandariza texto para datos de incidencia delictiva:
    - Convierte a string si no lo es
    - Elimina acentos
    - Convierte a t√≠tulo (primera letra may√∫scula)
    - Limpia caracteres especiales excepto n√∫meros y letras
    - Normaliza espacios
    """
    if pd.isna(texto) or texto == '':
        return texto
    
    if not isinstance(texto, str):
        texto = str(texto)
    
    # Normalizar acentos y caracteres especiales
    texto = texto.replace('√°', 'a').replace('√©', 'e').replace('√≠', 'i').replace('√≥', 'o').replace('√∫', 'u')
    texto = texto.replace('√Å', 'A').replace('√â', 'E').replace('√ç', 'I').replace('√ì', 'O').replace('√ö', 'U')
    texto = texto.replace('√±', 'n').replace('√ë', 'N')
    
    # Limpiar caracteres especiales pero mantener algunos importantes
    # Mantener letras, n√∫meros, espacios, par√©ntesis, guiones y algunos s√≠mbolos comunes en delitos
    texto = re.sub(r'[^a-zA-Z0-9\s\(\)\-\./]', '', texto)
    
    # Normalizar espacios m√∫ltiples
    texto = ' '.join(texto.strip().split())
    
    # Convertir a formato t√≠tulo
    return texto.title()

# Aplicar limpieza a columnas categ√≥ricas de texto
columnas_texto = ['Entidad', 'Bien_Juridico_Afectado', 'Tipo_Delito', 'Subtipo_Delito', 'Modalidad', 'Mes']

print("üîÑ Estandarizando texto en columnas categ√≥ricas:")

for columna in columnas_texto:
    print(f"\nüìù Procesando: {columna}")
    
    # Mostrar algunos valores antes
    valores_antes = df_copy[columna].cat.categories[:5].tolist()
    print(f"  Antes (muestra): {valores_antes}")
    
    # Convertir temporalmente a string para limpiar
    df_copy[columna] = df_copy[columna].astype('str')
    df_copy[columna] = df_copy[columna].apply(estandarizar_texto_delictivo)
    
    # Reconvertir a categorical
    df_copy[columna] = df_copy[columna].astype('category')
    
    # Mostrar algunos valores despu√©s
    valores_despues = df_copy[columna].cat.categories[:5].tolist()
    print(f"  Despu√©s (muestra): {valores_despues}")
    print(f"  Categor√≠as √∫nicas: {df_copy[columna].nunique()}")

print(f"\n‚úÖ Limpieza de texto completada")

# Verificar la calidad de la limpieza
print(f"\nüìä Resumen despu√©s de limpieza:")
for col in columnas_texto:
    print(f"  {col}: {df_copy[col].nunique()} categor√≠as √∫nicas")

# Mostrar ejemplos de datos limpios
print(f"\nüëÅÔ∏è Muestra de datos despu√©s de limpieza:")
muestra_limpia = df_copy[['Entidad', 'Tipo_Delito', 'Subtipo_Delito', 'Modalidad']].head(3)
print(muestra_limpia)
````

## 3. Regresi√≥n Lineal Simple

Aplicamos una regresi√≥n lineal simple para predecir la variable `Generacion_Estimada` en funci√≥n de una sola variable independiente. Se analiza su relaci√≥n con una gr√°fica de dispersi√≥n y se eval√∫a el modelo con m√©tricas como R¬≤, MSE y RMSE.


````python
# Usar todos los datos originales para regresi√≥n lineal simple
df['y'] = df['Incidencia_Delictiva']  # Ahora s√≠ existe y es num√©rica
display(df.head())
````

````python
# Solo continuar si la columna 'y' existe
if 'y' in df.columns:
    #2. Correlacion de Pearson
    def correlacion_pearson(x, y):
        X_mean = np.mean(x)
        Y_mean = np.mean(y)
        numerator = np.sum((x - X_mean) * (y - Y_mean))
        denominator = np.sqrt(np.sum((x - X_mean) ** 2) * np.sum((y - Y_mean) ** 2))
        return numerator / denominator
    #3. Calcular las ventas
    def calcular_parametros(x, y):
        x_mean = np.mean(x)
        y_mean = np.mean(y)
        beta1 = np.sum((x - x_mean) * (y - y_mean)) / np.sum((x - x_mean) ** 2)
        beta0 = y_mean - beta1 * x_mean
        return beta0, beta1
else:
    print("No se puede continuar: la columna 'y' no existe en el DataFrame agrupado.")
````

````python
# Agrupar por A√±o y Entidad para tener m√°s datos
df_grouped = df.groupby(['Anio', 'Entidad']).agg({'Incidencia_Delictiva': 'sum'}).reset_index()
df_grouped = df_grouped.rename(columns={'Incidencia_Delictiva': 'y'})

# Calcular correlaciones si existe 'y'
if 'y' in df_grouped.columns:
    correlaciones = {}
    columnas_numericas = df_grouped.select_dtypes(include=[np.number]).columns.drop('y', errors='ignore')

    for columna in columnas_numericas:
        corr = correlacion_pearson(df_grouped[columna], df_grouped['y'])
        correlaciones[columna] = abs(corr)

    if correlaciones:
        mejor_variable = max(correlaciones, key=correlaciones.get)
        print(f"Mejor candidato de X es: {mejor_variable}")
        print(f"Correlaci√≥n: {correlaciones[mejor_variable]:.4f}")
    else:
        print("No hay variables num√©ricas distintas a 'y' para correlacionar.")
else:
    print("No se puede calcular correlaciones: la columna 'y' no existe en el DataFrame.")
````

````python
# Resolver la ecuaci√≥n de la regresi√≥n solo si existen las variables necesarias
if 'mejor_variable' in locals() and 'y' in df_grouped.columns and mejor_variable in df_grouped.columns:
    x = df_grouped[mejor_variable].values
    y = df_grouped['y'].values
    beta0, beta1 = calcular_parametros(x, y)
    print(f"El valor de beta1 es: {beta1}")
    print(f"El valor de beta0 es: {beta0}")
else:
    print("No se puede resolver la regresi√≥n: faltan variables necesarias.")
````

````python
# Funci√≥n para realizar las predicciones
def hacer_prediccion(x_nuevo, beta0, beta1):
    return beta0 + beta1 * x_nuevo

# Realizar predicciones si todo est√° definido
if 'x' in locals() and 'beta0' in locals() and 'beta1' in locals():
    y_prediccion = hacer_prediccion(x, beta0, beta1)
    print("Valores reales de y:")
    print(y[:10])  # Mostrar solo los primeros 10 para no saturar
    print("Valores predichos de y:")
    print(y_prediccion[:10])  # Mostrar solo los primeros 10
else:
    print("No se puede hacer la predicci√≥n: faltan variables necesarias.")
````

````python
# Calcular r2 y mse solo si existen las variables necesarias
if 'y' in locals() and 'y_prediccion' in locals():
    def calcular_r2(y_real, y_pred):
        ss_res = np.sum((y_real - y_pred) ** 2)
        ss_tot = np.sum((y_real - np.mean(y_real)) ** 2)
        return 1 - (ss_res / ss_tot)

    def calcular_mse(y_real, y_pred):
        return np.mean((y_real - y_pred) ** 2)

    r2 = calcular_r2(y, y_prediccion)
    mse = calcular_mse(y, y_prediccion)
    print(f"valor de r2: {r2}")
    print(f"Valor de MSE: {mse}")
else:
    print("No se puede calcular r2 ni mse: faltan variables necesarias.")
````

````python
# --- Parte 1: b√∫squeda de K y entrenamiento final ---
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, f1_score, classification_report
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

k_range = range(1, 16)
weights_opts = ["uniform", "distance"]

results = []
for w in weights_opts:
    for k in k_range:
        knn = KNeighborsClassifier(n_neighbors=k, weights=w)
        knn.fit(X_train_knn, y_train_knn)
        y_pred = knn.predict(X_test_knn)
        acc = accuracy_score(y_test_knn, y_pred)
        f1m = f1_score(y_test_knn, y_pred, average="macro")
        results.append({"k": k, "weights": w, "accuracy": acc, "f1_macro": f1m})

res_df = pd.DataFrame(results)

# Curvas: Accuracy y F1-macro vs K
plt.figure(figsize=(9,4))
for w in weights_opts:
    sub = res_df[res_df["weights"] == w]
    plt.plot(sub["k"], sub["accuracy"], marker="o", label=f"Accuracy ({w})")
    plt.plot(sub["k"], sub["f1_macro"], marker="s", linestyle="--", label=f"F1-macro ({w})")
plt.xlabel("Valor de K")
plt.ylabel("M√©trica")
plt.title("Selecci√≥n de K y esquema de pesos")
plt.xticks(list(k_range))
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# Elegir mejor configuraci√≥n por F1-macro (y en empate por Accuracy)
best_row = res_df.sort_values(["f1_macro", "accuracy"], ascending=False).iloc[0]
best_k = int(best_row["k"])
best_w = best_row["weights"]
print(f"Mejor configuraci√≥n -> k={best_k}, weights='{best_w}' | "
      f"F1-macro={best_row['f1_macro']:.3f}, Acc={best_row['accuracy']:.3f}")

# Entrenar modelo final y evaluar
best_knn = KNeighborsClassifier(n_neighbors=best_k, weights=best_w)
best_knn.fit(X_train_knn, y_train_knn)
y_pred_knn = best_knn.predict(X_test_knn)

acc = accuracy_score(y_test_knn, y_pred_knn)
f1m = f1_score(y_test_knn, y_pred_knn, average="macro")
print(f"\nDesempe√±o final -> Accuracy: {acc:.3f} | F1-macro: {f1m:.3f}\n")
print("Reporte de clasificaci√≥n:\n", classification_report(y_test_knn, y_pred_knn))

# Guardamos cosas que usaremos en las partes 2 y 3
labels = np.sort(y_test_knn.unique())
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test_knn, y_pred_knn, labels=labels)
````

- **KNN** resuelve un problema distinto (clasificaci√≥n por **tipo de delito**) y se eval√∫a con **Accuracy**, **Precision**, **Recall** y **F1-macro** (esta √∫ltima es clave con clases desbalanceadas).

---

### An√°lisis Cr√≠tico de Fortalezas y Debilidades

| Modelo                    | Fortalezas                                                                 | Debilidades                                                                                 |
|---------------------------|-----------------------------------------------------------------------------|----------------------------------------------------------------------------------------------|
| Regresi√≥n Lineal Simple   | Muy interpretable; visualizaci√≥n directa (*scatter + l√≠nea*)               | Limitada a un √∫nico predictor; capta mal no-linealidades                                     |
| Regresi√≥n Lineal M√∫ltiple | Incorpora m√°s factores; mejor ajuste potencial                             | Riesgo de multicolinealidad; requiere preprocesamiento (dummies, escalado opcional)         |
| Clasificador KNN          | No asume forma funcional; f√°cil de implementar; capta patrones locales     | Muy sensible al **escaleo** y al **desbalance**; costo crece con datos; elecci√≥n de *k* clave |

---

### Recomendaciones para Uso (Contexto de Incidencia Delictiva)

- **Regresi√≥n Lineal Simple**: √∫til para una lectura r√°pida de tendencia con un solo factor (p. ej., evoluci√≥n por **A√±o**).  
- **Regresi√≥n Lineal M√∫ltiple**: recomendable para reportes estrat√©gicos cuando se incorporan variables categ√≥ricas (p. ej., **Entidad**), ofreciendo una explicaci√≥n m√°s rica del comportamiento de la **incidencia delictiva**.  
- **KNN (clasificaci√≥n por Tipo de Delito)**: viable para **etiquetar** registros cuando hay buenas features; si el dataset est√° desbalanceado, conviene aplicar **rebalanceo (SMOTE/undersampling)**, **tuning** y/o **PCA** antes de KNN, o considerar **√°rboles/ensembles**.  

---

### Conclusiones

- Cada modelo responde a objetivos distintos: **regresiones** para valores continuos (incidencia total) y **KNN** para **clases** (tipo de delito).  
- En nuestros datos, la m√∫ltiple ofrece una mejora frente a la simple, pero su poder predictivo depende de la calidad de variables (y puede ser modesto).  
- El KNN mostr√≥ **accuracy/F1-macro bajos** por **alto desbalance** y la naturaleza multiclase; se recomienda rebalanceo y/o modelos alternativos.  
- En conjunto, estos enfoques brindan una **visi√≥n integral**: tendencia y magnitud (regresiones) + clasificaci√≥n operativa (KNN).
